## Functions

```python
# Function to calcuate the probability of a single specific
# outcome across one roll of a num_of_sides-sided die
def f1Die1Outcome(num_of_sides):
    prob = 1 / num_of_sides
    prob = "{:.2%}".format(prob)
    probMessage = 'The probability of a single specific outcome on a {}-sided die is {}'.format(num_of_sides, prob)
    return probMessage

# Function to calculate the probability of obtaining a
# single specific outcome across num_of_rolls of num_of_sides-sided dice
# I.E. this tells you the probability of rolling ANY single 
# outcome, like 1, 4, 6, 12, etc.
def fNDice1Outcome(num_of_rolls, num_of_sides):
    prob = 1 - ((num_of_sides - 1) / num_of_sides)**num_of_rolls
    prob = "{:.2%}".format(prob)
    num_of_rolls = inflect.engine().number_to_words(num_of_rolls)
    probMessage = 'The probabiity of achieving a single specific outcome across {} {}-sided dice is {}'.format(num_of_rolls, num_of_sides, prob)
    return probMessage

# Function to calculate the probability of obtaining an outcome of
# desired_outcome or greater on at least one of num_of_rolls num_of_sides-sided dice
def fNDice1OutcomeGreater(num_of_rolls, num_of_sides, desired_outcome):
    prob = 1 - ((num_of_sides - (num_of_sides - desired_outcome + 1)) / num_of_sides)**num_of_rolls
    prob = "{:.2%}".format(prob)
    num_of_rolls = inflect.engine().number_to_words(num_of_rolls)
    probMessage = 'The probability of obtaining a(n) {} or greater across {} {}-sided dice is {}'.format(desired_outcome, num_of_rolls, num_of_sides, prob)
    return probMessage

# Function to calculate the probability of hitting a creature in D&D 5e
# using the creatures AC (vAC), your proficiency bonus (vProf), and your 
# ability score modifier (vASM)
def fProbToHit(vAC, vProf, vASM):
    prob = ((20 - (vAC - (vProf + vASM)))/100)*5
    prob = "{:.2%}".format(prob)
    probMessage = 'The probability of hitting a target with [AC {}] while having [Prof. +{}] and a [Ability mod. +{}] is {}'.format(vAC, vProf, vASM, prob)
    return probMessage

# Function to roll as many num_of_sides-sided dice as you want
def fRoll(num_of_rolls, num_of_sides):
    rolls = {}
    for i in range(1, num_of_rolls + 1):
        rolls[i] = r.randint(1, num_of_sides)
    nSum = sum(rolls.values())
    rolls['total'] = nSum
    return rolls
```
